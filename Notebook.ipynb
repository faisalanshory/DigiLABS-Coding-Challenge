# **Data Wrangling**

```python
#Read Excel File
df = pd.read_excel("/content/SocMed_Geographic.xlsx", sheet_name="Sheet1")
df.head(10)
```

```python
def extraword_remover(dfs):
  column_count = len(dfs.columns)
  row_count = len(dfs.index)
  for i in range(1, column_count):
    for j in range(row_count):
      if pd.isna(dfs.iloc[j,i]):
        continue
      elif "(" in dfs.iloc[j,i]:
        dfs.iloc[j,i] = dfs.iloc[j,i][0:-11]
  print('Extrawords in user number are removed succesfully')

extraword_remover(df)
```

```python
def whitespace_remover(dfs):
  column_count = len(dfs.columns)
  row_count = len(dfs.index)
  for i in range(column_count):
    for j in range(row_count):
      if pd.isna(dfs.iloc[j,i]):
        continue
      else:
        dfs.iloc[j,i] = dfs.iloc[j,i].strip()
        dfs.iloc[j,i] = dfs.iloc[j,i].strip('\u202c')
  print('Whitespaces in all cells are removed succesfully')

whitespace_remover(df)
```

```python
#Change Header to Country Name
df.at[0,'Social Media Platform']='Social Media Platform'
df.columns = df.iloc[0]
df = df[1:]

df.head(3)
```

```python
def comma_remover(dfs):
  column_count = len(dfs.columns)
  row_count = len(dfs.index)
  for i in range(1, column_count):
    for j in range(row_count):
      if pd.isna(dfs.iloc[j,i]):
        continue
      elif "," in dfs.iloc[j,i]:
        dfs.iloc[j,i] = re.sub(',', '', dfs.iloc[j,i])
  print('Commas in all user number are removed succesfully')

comma_remover(df)
```

```python
def space_remover(dfs):
  column_count = len(dfs.columns)
  row_count = len(dfs.index)
  for i in range(1, column_count):
    for j in range(row_count):
      if pd.isna(dfs.iloc[j,i]):
        continue
      else:
        dfs.iloc[j,i] = dfs.iloc[j,i].replace(' ','')
  print('Extraspaces in all user number are removed succesfully')

space_remover(df)
```

```python
def float_converter(dfs):
  column_name = len(dfs.columns)
  row_count = len(dfs.index)
  for i in range(1, column_name):
    for j in range(row_count):
      if isinstance(dfs.iloc[j,i], str):
        dfs.iloc[j,i] = float(dfs.iloc[j,i])
  print('All number are converted to float succesfully')     

float_converter(df)
```
